```python
import numpy as np
#Creating List
Salary = [120000,np.nan,56000,78000,45000,67000,np.nan]
print(Salary)
Incentive = [300,400,500,600,700]
print(Incentive)
print('Net Salary',Salary+Incentive)
```

```python
#Changing the list to series
import pandas as pd
salary_series=pd.Series(Salary)
incentive_series=pd.Series(Incentive)
print("Salary Series",salary_series)
print("Incentive Series",incentive_series)
```

```python
#Operations on Series
import pandas as pd
salary_series=pd.Series(Salary)
incentive_series=pd.Series(Incentive)
print("Salary Series",salary_series)
print("Incentive Series",incentive_series)
print("add:\n",salary_series.add(incentive_series))
print("sub:\n",salary_series.sub(incentive_series))
print("add:\n",salary_series.mul(incentive_series))
```

```python
# combine the  incentive series
print("Combination of both series:/n",salary_series.combine_first(incentive_series))
```

```python
#mapping of the dataframe
df2=pd.DataFrame(
    [('carrot','red',1),
     ('papaya','yellow',0),
     ('mango','yellow',0),
     ('apple','yellow',0),
     ],
    columns=['specifies','color','type']
    )
print("dataframe before mapping:")
print(df2)
mappings = {
    'carrot': 'veg',
    'papaya': 'fruit'
}
df2['type_name']=df2['specifies'].map(mappings)
print("Dataframe after mapping:")
print(df2)
```
